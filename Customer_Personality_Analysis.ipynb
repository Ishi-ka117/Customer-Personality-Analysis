**Missing values handled**

```python
print(csv.isnull().sum())
```

```python
csv['Income'] = csv['Income'].fillna(csv['Income'].median())
```

**Remove Duplicate Rows**

```python
print("Duplicate rows before cleaning:", csv.duplicated().sum())
csv = csv.drop_duplicates()
print("Duplicate rows after cleaning:", csv.duplicated().sum())
```

**Convert Dt_Customer to datetime**

```python
csv['Dt_Customer'] = pd.to_datetime(csv['Dt_Customer'], format='%d-%m-%Y', errors='coerce')
print("Data type of Dt_Customer:", csv['Dt_Customer'].dtype)
```

**standardize Column Names (lowercase + underscores)**

```python
csv.columns = csv.columns.str.strip().str.lower().str.replace(' ', '_')
print(csv.columns.tolist())
```

**Clean education and marital_status values**

```python
csv['education'] = csv['education'].str.strip().str.title()
csv['marital_status'] = csv['marital_status'].str.strip().str.title()
```

```python
print("Education values:", csv['education'].unique())
print("Marital status values:", csv['marital_status'].unique())
```

**Saving the Cleaned File**

```python
csv.to_csv("/content/drive/MyDrive/Elevate Labs/marketing_campaign_cleaned.csv", index=False)
print("Cleaned CSV saved successfully.")
```
